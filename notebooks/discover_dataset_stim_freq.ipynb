```python
import os
import glob
import pandas as pd
```

```python
def discover_dataset_stim_freq(folder):
    # Find all *_events.tsv files recursively
    event_files = glob.glob(
        os.path.join(folder, "**", "*_events.tsv"), recursive=True
    )

    all_unique_frequencies = set()
    for file in event_files:
        try:
            df = pd.read_csv(file, sep="\t")
            if "trial_type" in df.columns:
                unique_trial_types = df["trial_type"].unique()
                freqs = []
                for trial_type in unique_trial_types:
                    freqs.append(trial_type.split("_")[2])
                all_unique_frequencies.update(freqs)
        except Exception as e:
            print(f"Could not process file {file}: {e}")

    # Convert to sorted list for readability
    all_unique_frequencies = sorted(list(all_unique_frequencies))
    print(all_unique_frequencies)
```

```python
# Frequencies in Minipigs 2026:
minipigs_folder = (
    "/network/projects/neuro-galaxy/data/raw/neurosoft_minipigs_2026"
)
discover_dataset_stim_freq(minipigs_folder)

# Frequencies in Monkeys 2026:
monkeys_folder = (
    "/network/projects/neuro-galaxy/data/raw/neurosoft_monkeys_2026"
)
discover_dataset_stim_freq(monkeys_folder)
```
